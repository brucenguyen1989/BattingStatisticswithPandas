```python
import pandas as pd
import numpy as np
```

__1) Load in the appropriate csv file as a pandas dataframe (batting.csv)__

```python
batting = pd.read_csv('batting.csv')
```

__2)	Print out the dimensions and info about the dataframe you just created__

```python
print(batting.shape)
```

__3) How many players have hit 40 or more HRs in one single season? (Number only)__

```python
q3 = batting.groupby(['playerID','yearID']).agg({'HR':np.sum})
q3[q3['HR']>=40].reset_index().nunique()[0]
```

__4)	How many players have hit 600 or more HRs for their career? (Dataframe with name and player first and last name only and amount of HRs)__

```python
q4b = batting.groupby(['playerID','nameFirst','nameLast']).agg({'HR':np.sum}).sort_values('HR',ascending = False)
q4a = q4b[q4b['HR']>600].reset_index()
q4 = pd.DataFrame({'FirstName':q4a['nameFirst'],'LastName':q4a['nameLast'],'HR':q4a['HR']}).reindex(['FirstName','LastName','HR'],axis=1)
q4
```

__5)	How many unique players have hit 40 2Bs or more, 10 3Bs or more, 200 Hits or more, and 30 HRs or more in one season? (Number Only)__

```python
q5a = batting.groupby(["playerID","yearID"]).agg({'2B':np.sum,'3B':np.sum,'H':np.sum,'HR':np.sum})
q5 = q5a[(q5a['2B']>=40) & (q5a['3B']>=10) & (q5a['H']>=200) & (q5a['HR']>=30)]
q5.reset_index().nunique()[0]
```

__6)	How many players seasons have had 100 or more SBs? (Dataframe, first name, last name, yearID, number of SBs, ordered from descending SBs )__

```python
q6b = batting.groupby(['playerID','nameFirst','nameLast','yearID']).agg({'SB':np.sum}).sort_values('SB',ascending=False)
q6a = q6b[q6b['SB']>=100].reset_index()
q6 = pd.DataFrame({'FirstName':q6a['nameFirst'],'LastName':q6a['nameLast'],'yearID':q6a['yearID'],'SB':q6a['SB']}).reindex(['FirstName','LastName','yearID','SB'],axis=1)
q6
```

__10)	Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe, first name, last name, OBP, ordered descending by OBP)__

```python
batting.fillna(value=0,inplace = True)
```

```python
batting['OBP'] = ((batting['H']+batting['BB']+batting['HBP'])/(batting['AB']+batting['BB']+batting['HBP']+batting['SF']))
batting['PA'] = (batting['AB']+batting['BB']+batting['HBP']+batting['SF']+batting['SH'])
batting['OBP'] = pd.Series(["{0:.2f}%".format(val * 100) for val in batting['OBP']])
```

```python
q10a = batting[batting['PA']>=500].sort_values('OBP',ascending=False).head(5).reset_index()
q10 = pd.DataFrame({'FirstName':q10a['nameFirst'],'LastName':q10a['nameLast'],'OBP':q10a['OBP']})
q10
```

__13)	Who led MLB with the most hits the most times?  And how many times?  (Dataframe, Number of hits)__

__23)	Who had more than the league average HRs (for players with 500 or more ABs) in 2012 (filter out all players with less 500 ABs)? (Dataframe first name, last name, HR descending)__	

```python
q23c = batting[(batting['yearID']==2012)].groupby(['lgID']).agg({'HR':np.mean}).reset_index()
lg_average = pd.DataFrame({'lgID':q23c['lgID'],'HR_average':q23c['HR']})
lg_average
```

```python
q23b = batting[(batting['AB']>=500)&(batting['yearID']==2012)]
q23a = q23b.merge(lg_average,on='lgID')
q23x = q23a[q23a['HR']>q23a['HR_average']].sort_values('HR',ascending = False)
q23 = pd.DataFrame({'FirstName':q23x['nameFirst'],'LastName':q23x['nameLast'],'HR':q23x['HR']})
q23.reset_index().drop(['index'],axis = 1)
```

__24)	 Who is the youngest player to hit 50 or more HRs in a single season? (Dataframe, first name, last name, HRs, season)__

__25)	Who are the five youngest players to hit 300 or more HRs for their career? (Dataframe, first name, last name, season they eclipsed more than 300 HRs)__

```python
q25_1 = batting.groupby(['playerID','nameFirst','nameLast','yearID','Age']).agg({'HR':'sum'}).reset_index()
q25_1['Cum_HR'] = q25_1.groupby(['playerID'])['HR'].cumsum()
q25_2 = q25_1[q25_1['Cum_HR']>=300].reset_index().drop(['index'],axis=1)
q25_3 = q25_2.groupby('playerID').apply(lambda x:x[x["Age"]==x["Age"].min()]).sort_values('Age').head(5)
q25_4 = pd.DataFrame({'FirstName':q25_3['nameFirst'],'LastName':q25_3['nameLast'],'Age':q25_3['Age'],'Season':q25_3['yearID'],'Cum_HR':q25_3['Cum_HR']})
q25_5 = q25_4.reset_index().sort_values('Season').drop(['playerID','level_1'],axis=1)
q25_5.reset_index().drop(['index'],axis=1)
```
